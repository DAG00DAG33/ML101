```
from sklearn.decomposition import PCA

def reduce_size(vecs, train = False):
    if (train):
        reduce_size.pca = PCA(n_components=2)
        reduced = reduce_size.pca.fit_transform(vecs)
    else:
        reduced = reduce_size.pca.transform(vecs)
    return reduced
```
